```python
df_monthly_sale =df.groupby('month_year').sum()['Total'].reset_index()
df_monthly_sale
```

```python
plt.figure(figsize=(15,6)) 
plt.plot(df_monthly_sale['month_year'],df_monthly_sale['Total'], color='red')
plt.xticks(rotation = 'vertical', size=8)
plt.show()
```

```python
df_daily_sale =df.groupby('month_year').sum()['Total'].reset_index()
df_daily_sale
```

```python
plt.figure(figsize=(20,6)) 
plt.plot(df_daily_sale['month_year'],df_daily_sale['Total'], color='red')
plt.xticks(rotation = 'vertical', size=8)
plt.show()
```

# <h2> which are the top products by sales?

```python
# Grouping product name column

prod_lines = pd.DataFrame(df.groupby('Product_line').sum()['Total'])
```

```python
#sorting prod_sales column
prod_lines =prod_lines.sort_values('Total',ascending=False)
prod_lines
```

# which are the most selling product?

```python
prod_quantity=pd.DataFrame(df.groupby('Product_line').sum()['Quantity'])
```

```python
prod_quantity= prod_quantity.sort_values('Quantity',ascending=False)
prod_quantity
```

# Type of payment

```python
payment_mode= pd.DataFrame(df.groupby('Payment').count()['Invoice_ID'])
payment_mode
```

```python
payment_mode= payment_mode.sort_values('Invoice_ID',ascending=False)
payment_mode
```

```python
plt.figure(figsize=(10,8))
sns.countplot(df['Payment'])
plt.show()
```
